# Trees

## Binary tree & Tree traversal

```python
class Node:
    data = None
    left = None
    right = None
    
    def __init__(self):
        self.data = None 
        self.left = None
        self.right = None
        return

    def __init__(self, data):
        self.data = data 
        self.left = None
        self.right = None
        return

    def addLeft(self, node):
        self.left = node
        return
    
    def addRight(self, node):
        self.right = node
        return
    
root = Node(5)
node1 = Node(4)
node2 = Node(7)
node3 = Node(8)
node4 = Node(1)


node1.left = node3
node1.right = node4
root.left = node1
root.right = node2

# preorder
def preorder(node):
    if node is None:
        return
    print(node.data)
    preorder(node.left)
    preorder(node.right)
    return

# inorder
def inorder(node):
    if node is None:
        return
    inorder(node.left)
    print(node.data)
    inorder(node.right)
    return

# postorder
def postorder(node):
    if node is None:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.data)
    return

print('Pre-order')
preorder(root)

print('In-order')
inorder(root)


print('Post-order')
postorder(root)
```

## Binary search tree

```python
class BST:
    root = None
    
    def __init(self):
        root = None
    
    def insert(self, data):
        new_node = Node(data)
        if self.root is None:
            self.root = new_node
        else:
            self.insert_inorder(self.root, new_node)
        return
    
    def insert_inorder(self, node, new_node):
        if new_node.data < node.data:
            if node.left == None:
                node.left = new_node
                return
            else:
                self.insert_inorder(node.left, new_node)
        elif new_node.data > node.data:
            if node.right == None:
                node.right = new_node
                return
            else:
                self.insert_inorder(node.right, new_node)
        return
    
    def search(self, target):
        return self.search(self.root, target)
    
    def search(self, node, target):
        if node.data == target:
            return True
        if target < node.data:
            if node.left is not None:
                return self.search(node.left, target)
        else:
            if node.right is not None:
                return self.search(node.right, target)
    return False


class Node:
    data = None
    left = None
    right = None
    
    def __init__(self):
        self.data = None 
        self.left = None
        self.right = None
        return

    def __init__(self, data):
        self.data = data 
        self.left = None
        self.right = None
        return

    def addLeft(self, node):
        self.left = node
        return
    
    def addRight(self, node):
        self.right = node
        return
    
bst = BST()
bst.insert(5)
bst.insert(3)
bst.insert(8)
bst.insert(2)
bst.insert(1)
bst.insert(7)
bst.insert(10)

# inorder
def inorder(node):
    if node is None:
        return
    inorder(node.left)
    print(node.data)
    inorder(node.right)
    return

inorder(bst.root)

print()
bst.insert(4)
inorder(bst.root)
```
